```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
mouse_metadata
```

```python
#start cleaning data & break it down 
study_results = pd.read_csv(study_results_path)
study_results.head()
```

```python
#put data frames together 
bigdata_df = pd.merge(study_results, mouse_metadata, on=["Mouse ID"])
bigdata_df.head()


```

```python


# Create a clean DataFrame by dropping the duplicate mouse by its ID.

bigdata_df_fixed = bigdata_df[bigdata_df['Mouse ID'] != "g989"]
bigdata_df_fixed = pd.DataFrame(bigdata_df_fixed)
bigdata_df_fixed
```

```python
# Checking the number of mice in the clean DataFrame.
Mice_count_2 = len(pd.unique(bigdata_df_fixed['Mouse ID']))
  
Mice_count_2
```

## Summary Statistics

```python
#created a df to start breaking the data down with
starter_df = bigdata_df_fixed[['Drug Regimen', 'Tumor Volume (mm3)']]
starter_df.head()
```

```python
#calculated the mean 
avgvol_df  = starter_df.groupby("Drug Regimen").mean()
avgvol_df = pd.DataFrame(avgvol_df)

#renamed Columns
avgvol_df = avgvol_df.rename(columns = {'Tumor Volume (mm3)' : 'Mean Volume'})
avgvol_df.head(3)
```

```python
#calculated the mean
medvol_df  = starter_df.groupby("Drug Regimen").median()
medvol_df = pd.DataFrame(medvol_df)
#renamed columns
medvol_df = medvol_df.rename(columns = {'Tumor Volume (mm3)' : 'Median Volume'})
medvol_df.head(3)
```

```python
#Calculated Variance & renamed Columns
variance_df = starter_df.groupby("Drug Regimen").var()
variance_df = pd.DataFrame(variance_df)

variance_df = variance_df.rename(columns = {'Tumor Volume (mm3)' : 'Variance'})
variance_df.head(3)
```

```python
#Calculated Standard Deviation & renamed Columns
standev_df = starter_df.groupby("Drug Regimen").std()
standev_df = pd.DataFrame(standev_df)

standev_df = standev_df.rename(columns = {'Tumor Volume (mm3)' : 'Std. Deviation'})
standev_df.head(3)
```

```python
#Calculated Standard Error & renamed Columns
error_df = starter_df.groupby("Drug Regimen").sem()
error_df = pd.DataFrame(error_df)
                        
error_df = error_df.rename(columns = {'Tumor Volume (mm3)' : 'Std. Error'})                      
error_df.head(3)
```

```python
#began merging DF's 
df_1 = pd.merge(avgvol_df, medvol_df, on="Drug Regimen")
df_1


```

```python
#Continued merging Df's
df_2 = pd.merge(df_1, variance_df, on="Drug Regimen")
df_2 
```

```python
#Continued merging Df's
df_3 = pd.merge(df_2, standev_df, on="Drug Regimen")
df_3
```

```python
#generate stats table summary 
df_plot = pd.merge(df_3, error_df, on="Drug Regimen")
df_plot
```

```python
#Aggregate summary
aggregate_df = starter_df.groupby("Drug Regimen").agg(["mean", "median", "var", "std", "sem"])
aggregate_df
```
